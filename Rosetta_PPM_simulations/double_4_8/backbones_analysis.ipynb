```python
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
import glob
import os
from shutil import copyfile
import seaborn as sns
%matplotlib inline
```

```python
for folder_id in range(0,10):
    score_file = str(folder_id) + "/" + str(folder_id) + "__score.sc"
    dsc = pd.read_csv(score_file, sep='\s+', header=1)
    dclashes = dsc.loc[dsc['vdw'] < 1.0]
    for model in dclashes.description:
        source = str(folder_id) + "/" + model + ".pdb"
        target = "best/" + model + ".pdb"
        copyfile(source, target)
```

```python
for folder_id in range(0,10):
    score_file = str(folder_id) + "/" + str(folder_id) + "__score.sc"
    dsc = pd.read_csv(score_file, sep='\s+', header=1)

    dclashes = dsc.loc[(dsc['vdw'] < 1.0) & (dsc['omega'] < 21.0) & (dsc['cen_rama'] < 5.0)]
    print(dclashes.shape)
    for model in dclashes.description:
        source = str(folder_id) + "/" + model + ".pdb"
        target = "best_40/" + model + ".pdb"
        copyfile(source, target)
```
